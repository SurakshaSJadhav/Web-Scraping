## Accessing table using wikitable

```python
#Reading Wiki tables
import pandas as pd
```

```python
# change the index table
from pandas.io.html import read_html
import urllib.request

class AppURLopener(urllib.request.FancyURLopener):
    version = "Mozilla/5.0"
opener = AppURLopener()
page = opener.open('https://www.worldometers.info/coronavirus/')

wikitables = read_html(page,attrs={"id":"main_table_countries_today"})

print ("Extracted {num} wikitables".format(num=len(wikitables)))
```

## Accessing the table using BeautifulSoup

```python
import csv
from urllib.request import urlopen
from bs4 import BeautifulSoup
import urllib.request
```

```python
class AppURLopener(urllib.request.FancyURLopener):
    version = "Mozilla/5.0"
opener = AppURLopener()
```

```python
html = opener.open("https://www.worldometers.info/coronavirus/")
```

```python
soup = BeautifulSoup(html, "html.parser")
table = soup.findAll("table", {"id":"main_table_countries_today"})[0]
rows = table.findAll("tr")
```

```python
with open("Covid19cases.csv", "wt+", newline="") as f:
    writer = csv.writer(f)
    for row in rows:
        csv_row = []
        for cell in row.findAll(["td", "th"]):
            csv_row.append(cell.get_text())
        writer.writerow(csv_row)
```
